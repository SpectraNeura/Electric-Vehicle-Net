```python
train_count = len(train)
combined = pd.concat([train, test])
combined.shape, train_count
```

```python
combined = combined.drop('ID', axis=1)
combined
```

```python
# 배터리용량 예측을 위한 데이터 라벨 인코딩

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = combined.select_dtypes(include=object).columns
for col in cols:
  combined[col] = le.fit_transform(combined[col])
combined.head()
```

```python
cond1 = combined['배터리용량'].isnull()
preprocess_combined = combined[~cond1]
preprocess_combined
```

```python
preprocess_combined.corr(numeric_only=True)
```

# RandomForest Regressor()로 배터리용량 예측하여 완전한 데이터로 변환

```python
# target
target = preprocess_combined.pop('배터리용량')

from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(preprocess_combined, target, test_size=0.2, random_state=0)
print(X_tr.shape,X_val.shape,y_tr.shape,y_val.shape)
```

```python
# 랜덤포레스트로 전처리 모델 생성
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error as rmse
rf = RandomForestRegressor(random_state=0, n_estimators=200, max_depth=25)
rf.fit(X_tr,y_tr)
pred_pre_rf = rf.predict(X_val)
rmse(y_val, pred_pre_rf)

# baseline = 1.859323887774522
# n_estimators=200, 1.8565866718213786
# n_estimators=300, 1.8553501955071392
```

```python
# LightGBM로 전처리 모델 생성
import lightgbm as lgb
lg = lgb.LGBMRegressor(random_state=0, verbose=-1, n_estimators=100, max_depth=9)
lg.fit(X_tr,y_tr)
pred_pre_lgbm = lg.predict(X_val)
rmse(y_val, pred_pre_lgbm)
# baseline = 1.9677652825969485
# n_estimators=200, 1.969834708134316
# max_depth=9, 1.9590362410379394
```

# 전처리 모델은 randomforest로 진행

```python
# train 라벨 인코딩 데이터 추출
train = combined[:train_count]
# test 결과 제출용 ID 추출
test_id = test.pop('ID')
# test 라벨 인코딩 데이터 추출
test = combined[train_count:]

train.shape,test.shape
```
